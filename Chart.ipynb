```python
import matplotlib.pyplot as plt

groups=["6-8","9-11","12-14","15-17"]
students=[40,50,30,20]

plt.bar(groups, students, color="red")
plt.title("bar plot", fontsize=14)
plt.xlabel("groups", fontsize=12)
plt.ylabel("students", fontsize=12)
plt.show()
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns
data = { "Category": ['A', 'A', 'A', 'B', 'В', 'B', 'С', 'C', 'C'],
"Values": [10, 12, 15, 20, 22, 25, 30, 35, 40] }
sns.boxplot (x='Category',y='Values', data=data)
plt.show()
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns

tips = sns.load_dataset('tips')
sns.boxplot(x="day" , y="total_bill" , data=tips )

plt.title("title", fontsize=14)
plt.show()
```
